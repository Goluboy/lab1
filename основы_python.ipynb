#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(text: str) -> int:
    vowels = {'a', 'e', 'i', 'o', 'u'}
    cnt = 0
    for char in text.lower():
        if char in vowels:
            cnt += 1
    return cnt
```

```python
class TestCountVowels(unittest.TestCase):
    def test_empty_string(self):
        self.assertEqual(count_vowels(""), 0)

    def test_only_vowels(self):
        self.assertEqual(count_vowels("aeiou"), 5)

    def test_only_consonants(self):
        self.assertEqual(count_vowels("bcdfghjklmnpqrstvwxyz"), 0)

    def test_mixed_case_vowels(self):
        self.assertEqual(count_vowels("AeIoU"), 5)

    def test_mixed_content(self):
        self.assertEqual(count_vowels("Hello World"), 3)

    def test_uppercase_only(self):
        self.assertEqual(count_vowels("AEIOU"), 5)

    def test_single_character_vowel(self):
        self.assertEqual(count_vowels("a"), 1)

    def test_single_character_consonant(self):
        self.assertEqual(count_vowels("b"), 0)

    def test_no_vowels_in_sentence(self):
        self.assertEqual(count_vowels("Rhythm"), 0)

unittest.main(argv=["first-arg-is-ignored"], exit=False, verbosity=2)
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique_chars(s: str) -> bool:
    return len(set(s)) == len(s)
```

```python
import unittest

class TestHasUniqueCharacters(unittest.TestCase):
    def test_empty_string(self):
        self.assertTrue(is_unique_chars(""))

    def test_all_unique(self):
        self.assertTrue(is_unique_chars("abcdef"))

    def test_contains_duplicates(self):
        self.assertFalse(is_unique_chars("hello"))

    def test_case_sensitivity(self):
        self.assertTrue(is_unique_chars("AaBb"))

    def test_single_character(self):
        self.assertTrue(is_unique_chars ("z"))

    def test_all_identical(self):
        self.assertFalse(is_unique_chars("mmmm"))

    def test_spaces_and_punctuation(self):
        self.assertTrue(is_unique_chars("a b!@#"))
        self.assertFalse(is_unique_chars("a a"))

unittest.main(argv=["first-arg-is-ignored"], exit=False, verbosity=2)
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n: int) -> int:    
    steps = 0
    while n >= 10:
        product = 1
        temp = n
        while temp > 0:
            product *= temp % 10
            temp //= 10
        n = product
        steps += 1
        
    return steps
```

```python
class TestMultiplicationSteps(unittest.TestCase):
    def test_examples_from_task(self):
        self.assertEqual(magic(39), 3)
        self.assertEqual(magic(4), 0)
        self.assertEqual(magic(999), 4)

    def test_single_digit_numbers(self):
        for i in range(1, 10):
            self.assertEqual(magic(i), 0, f"Ожидалось 0 для числа {i}")

    def test_numbers_with_zero(self):
        self.assertEqual(magic(10), 1)
        self.assertEqual(magic(205), 1)
        self.assertEqual(magic(800), 1)

    def test_known_persistence_values(self):
        self.assertEqual(magic(25), 2)
        self.assertEqual(magic(679), 5)

unittest.main(argv=["first-arg-is-ignored"], exit=False, verbosity=2)
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError()
    if not pred:
        raise ValueError()

    n = len(pred)
    sum_squared_diff = sum((p - t) ** 2 for p, t in zip(pred, true))
    return sum_squared_diff / n
```

```python
class TestMSE(unittest.TestCase):
    def test_identical_vectors(self):
        self.assertAlmostEqual(mse([1, 2, 3], [1, 2, 3]), 0.0)

    def test_basic_case(self):
        self.assertAlmostEqual(mse([10, 15, 20], [20, 20, 20]), 41.666666666666664)

    def test_single_element(self):
        # (5 - 3)^2 / 1 = 4.0
        self.assertAlmostEqual(mse([5], [3]), 4.0)

    def test_negative_numbers(self):
        self.assertAlmostEqual(mse([-2, 3], [2, -3]), 26.0)

    def test_large_integers(self):
        val1 = [10**5, -10**5]
        val2 = [0, 0]
        expected = ((10**5)**2 + (-10**5)**2) / 2
        self.assertAlmostEqual(mse(val1, val2), expected)

    def test_different_lengths(self):
        with self.assertRaises(ValueError):
            mse([1, 2, 3], [1, 2])

    def test_empty_vectors(self):
        with self.assertRaises(ValueError):
            mse([], [])

unittest.main(argv=["first-arg-is-ignored"], exit=False, verbosity=2)
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def format_prime_factors(n: int) -> str:
    if not isinstance(n, int) or n < 2:
        raise ValueError()

    factors = []
    d = 2
    temp = n

    while d * d <= temp:
        if temp % d == 0:
            count = 0
            while temp % d == 0:
                count += 1
                temp //= d
            factors.append((d, count))
        d += 1

    if temp > 1:
        factors.append((temp, 1))

    return "".join(f"({p}**{exp})" if exp > 1 else f"({p})" for p, exp in factors)
```

```python
class TestFormatPrimeFactors(unittest.TestCase):
    def test_example_from_task(self):
        self.assertEqual(format_prime_factors(86240), "(2**5)(5)(7**2)(11)")

    def test_prime_number(self):
        self.assertEqual(format_prime_factors(17), "(17)")

    def test_power_of_prime(self):
        self.assertEqual(format_prime_factors(64), "(2**6)")

    def test_product_of_distinct_primes(self):
        self.assertEqual(format_prime_factors(30), "(2)(3)(5)")

    def test_smallest_valid_input(self):
        self.assertEqual(format_prime_factors(2), "(2)")

    def test_composite_with_mixed_exponents(self):
        self.assertEqual(format_prime_factors(12), "(2**2)(3)")
        self.assertEqual(format_prime_factors(18), "(2)(3**2)")
        self.assertEqual(format_prime_factors(100), "(2**2)(5**2)")

    def test_large_number(self):
        self.assertEqual(format_prime_factors(30030), "(2)(3)(5)(7)(11)(13)")

    def test_invalid_inputs(self):
        with self.assertRaises(ValueError):
            format_prime_factors(0)
        with self.assertRaises(ValueError):
            format_prime_factors(-42)
        with self.assertRaises(ValueError):
            format_prime_factors(1)
        with self.assertRaises(ValueError):
            format_prime_factors(3.14)
        with self.assertRaises(ValueError):
            format_prime_factors("123")

unittest.main(argv=["first-arg-is-ignored"], exit=False, verbosity=2)
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    if not isinstance(number, int) or number < 1:
        raise ValueError()

    k = 1
    current_sum = 0
    while current_sum < number:
        current_sum += k * k
        if current_sum == number:
            return k
        k += 1
        
    return "It is impossible"
```

```python
class TestFindPyramidLevel(unittest.TestCase):
    def test_valid_pyramids(self):
        self.assertEqual(pyramid(1), 1)
        self.assertEqual(pyramid(5), 2)
        self.assertEqual(pyramid(14), 3)
        self.assertEqual(pyramid(30), 4)
        self.assertEqual(pyramid(55), 5)
        self.assertEqual(pyramid(91), 6)

    def test_invalid_pyramids(self):
        self.assertEqual(pyramid(2), "It is impossible")
        self.assertEqual(pyramid(3), "It is impossible")
        self.assertEqual(pyramid(100), "It is impossible")
        self.assertEqual(pyramid(12345), "It is impossible")

    def test_large_number(self):
        self.assertEqual(pyramid(338350), 100)
        self.assertEqual(pyramid(333833500), 1000)

    def test_invalid_inputs(self):
        with self.assertRaises(ValueError):
            pyramid(0)
        with self.assertRaises(ValueError):
            pyramid(-15)
        with self.assertRaises(ValueError):
            pyramid(3.14)
        with self.assertRaises(ValueError):
            pyramid("55")

unittest.main(argv=["first-arg-is-ignored"], exit=False, verbosity=2)
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n: int) -> bool:
    if not isinstance(n, int) or n <= 0:
        raise ValueError("На вход должно быть подано положительное целое число.")

    digits = [int(d) for d in str(n)]
    length = len(digits)
    mid = length // 2

    if length % 2 == 0:
        left_part = digits[:mid - 1]
        right_part = digits[mid + 1:]
    else:
        left_part = digits[:mid]
        right_part = digits[mid + 1:]

    return sum(left_part) == sum(right_part)
```

```python
import unittest

class TestIsBalanced(unittest.TestCase):
    def test_single_and_double_digit(self):
        self.assertTrue(is_balanced(5))
        self.assertTrue(is_balanced(42))

    def test_balanced_odd_length(self):
        self.assertTrue(is_balanced(121))
        self.assertTrue(is_balanced(12321))
        self.assertTrue(is_balanced(90009))

    def test_balanced_even_length(self):
        self.assertTrue(is_balanced(1001))
        self.assertTrue(is_balanced(123321))
        self.assertTrue(is_balanced(556655))

    def test_unbalanced_odd_length(self):
        self.assertFalse(is_balanced(123))
        self.assertFalse(is_balanced(12345))
        self.assertFalse(is_balanced(900))

    def test_unbalanced_even_length(self):
        self.assertFalse(is_balanced(1234))
        self.assertFalse(is_balanced(1122))
        self.assertFalse(is_balanced(9981))

    def test_large_balanced_number(self):
        self.assertTrue(is_balanced(1234567654321))

    def test_invalid_inputs(self):
        with self.assertRaises(ValueError):
            is_balanced(0)
        with self.assertRaises(ValueError):
            is_balanced(-15)
        with self.assertRaises(ValueError):
            is_balanced(3.14)
        with self.assertRaises(ValueError):
            is_balanced("12321")

unittest.main(argv=["first-arg-is-ignored"], exit=False, verbosity=2)
```
